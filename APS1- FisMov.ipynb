```python
import math
from numpy import arange
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline

#definindo os parâmetros iniciais
O=math.radians(20)
g=10.0374912 #m/s**2
l=2.2 #m


#fórmula do odeint
def EqDif(Solucoes, t):
    o=Solucoes[0]
    Vo=Solucoes[1]
    dodt=Vo
    dVodt=-g*math.sin(o)/l
    return [dodt, dVodt]

#definimos o delta t
dt=0.001
#lista de tempo
t=arange(0,5,dt)

#condições iniciais
S0=[O,0] 

#executando o ODEINT
Solucao=odeint(EqDif, S0, t)
x=[]
y=[]
for i in range(1,len(Solucao[:,0])):
    o=(Solucao[:,0])[i]
    Vo=(Solucao[:,1])[i]
    x.append(l*math.cos(o))
    y.append(l*math.sin(o))
```

```python
#Plotando os resultados
plt.plot(y,x)
plt.axis([min(y), max(y), max(x), min(x)])
plt.title('Movimento do pêndulo')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
graus=[]
for i in range(0,len(Solucao[:,0])):
    angulo=math.degrees(Solucao[:,0][i])
    graus.append(angulo)



#plotando o gráfico do período
plt.plot(t,graus)
plt.title('Gráfico do Período do Pêndulo')
plt.ylabel('Ângulo')
plt.xlabel('Tempo(s)')
plt.grid(True)
plt.show()

```

```python
for i in range (0,len(graus)):
    if graus[i]==min(graus):
        períodográfico=2*(t[i])

períodofórmula=2*math.pi*((l/g)**(1/2))

    
        
```
